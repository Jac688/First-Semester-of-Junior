```python
textFile, y = readfile()
```

```python
import copy
```

```python
len(textFile)
```

```python
numFiles = len(textFile)
```

```python
# Removing puncuation
import re
remover_htmlLabel = re.compile('<.*?>')
remover_punctuation =  "\\【.*?】+|\\《.*?》+|\\#.*?#+|[.!/_,$&%^*()<>+""'?@|:~{}#]+|[——！\\\，。=？、：“”‘’￥……（）《》【】]"
```

```python
for i in range(numFiles):
    textFile[i][1] = re.sub(remover_htmlLabel,' ',textFile[i][1])
    textFile[i][1] = re.sub(remover_punctuation,' ',textFile[i][1])
```

```python
# Partition
textFile_splited = copy.deepcopy(textFile)
for i in range(numFiles):
    textFile_splited[i][1] = textFile[i][1].split()
```

```python
# Removing stopwords
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
filter_words = copy.deepcopy(textFile_splited)
for i in range(numFiles):
    filter_words[i][1] = [w for w in filter_words[i][1] if w not in stopwords.words('english')]
```

```python
# stemming and unify case
from nltk.stem import SnowballStemmer
stemmer = SnowballStemmer('english')
stemming_words = copy.deepcopy(filter_words)
for i in range(numFiles):
    stemming_words[i][1] = [stemmer.stem(w.lower()) for w in filter_words[i][1]]
```

```python
# Lemmatization and unify case
from nltk.stem import WordNetLemmatizer
wn = WordNetLemmatizer()
lemmatization_words = copy.deepcopy(filter_words)
for i in range(numFiles):
    lemmatization_words[i][1] = [wn.lemmatize(w.lower()) for w in filter_words[i][1]]
```

```python
for i in range(numFiles):
    stemming_words[i][1] = [' '.join(stemming_words[i][1])]
    lemmatization_words[i][1] = [' '.join(lemmatization_words[i][1])]
```

## Tokenization

**Methods:**
  1. Bag of Words (TF-IDF)
  2. N-gram

```python
# using lemmatization as example
corpus = []
for i in range(numFiles):
    corpus.append((lemmatization_words[i][1][0]))
```

```python
# TF-IDF
from sklearn.feature_extraction.text import TfidfVectorizer
vector = TfidfVectorizer()
tf_data = vector.fit_transform(corpus)
vector.vocabulary_
```

```python
import pandas as pd
df = pd.DataFrame(tf_data.toarray(), columns=vector.get_feature_names())
df
```

```python
# N-gram
from sklearn.feature_extraction.text import CountVectorizer
vector_bigram = CountVectorizer(min_df=1, ngram_range=(2,2))
X = vector_bigram.fit_transform(corpus)
vector_bigram.vocabulary_
```

```python
df = pd.DataFrame(X.toarray(), columns=vector_bigram.get_feature_names())
```
