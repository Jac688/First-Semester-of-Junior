```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
filename="./dataset/test.csv"

# import the test data 
from pandas import read_csv
f=open(filename,encoding='UTF-8')
test=read_csv(f)
print(test)
```

```python
filename="./dataset/training.csv"
f=open(filename,encoding='UTF-8')
training=read_csv(f)
print(training)
```

```python
# distance_1d = np.array([])

def knn(training, test, k):
    distance_1d = []
    res = []
    train_label = training['evaluation']
    train_attr = training.drop('evaluation', axis=1)
    # 
    for i in range(0, training.shape[0]):
        distance_1d = []
        for j in range(0, training.shape[0]):
            dis = (test.iloc[i]!=train_attr.iloc[j]).sum()
            distance_1d.append(dis)
#             print(distance_1d)
        distance_min = pd.DataFrame(distance_1d)
        distance_min.columns=['distance']
        
        min_dis = min(distance_1d)
        distance_min = distance_min.sort_values(by=['distance'], ascending=True)
        first_k = distance_min.iloc[:k, :]
        
# error         distance_max = distance_max[distance_max.iloc[:,0].max()]
        max_label = train_label[first_k.index]
#         print(first_k)
#         print(max_label)
        max_label_des   = max_label.value_counts()
        max_lable_count = max_label_des.max()
    
        
        max_label_res = max_label_des[max_label_des==max_lable_count].index[0]
        print(max_label_res)
        res.append(max_label_res)
#     print(res)
    return res
       
```

```python
knn_train_res = knn(training, training.drop(['evaluation'], axis=1), 3)
```

```python
knn_train_res = knn(training, test, 3)
```

```python
# analysis the result by train_train and train_test
# unacc as positive and acc as negative
knn_temp_train = pd.Series(knn_train_res)
knn_train = pd.Series(knn_train_res).value_counts()
print(knn_train)

knn_unacc_train = knn_train[0]/(knn_train[0]+knn_train[1])
knn_acc_train = knn_train[1]/(knn_train[0]+knn_train[1])
print(knn_unacc_train, knn_acc_train)

print("============================================")
train_res = pd.DataFrame([knn_train_res, training['evaluation']]).T
train_res.columns = ['pred', 'actl']
train_res_des = train_res.value_counts()


# score with acc is negative and unacc is positive
TP = train_res_des['unacc']['unacc']
FN = train_res_des['acc']['unacc']
FP = train_res_des['unacc']['acc']
TN =  train_res_des['acc']['acc']
precision = TP/(TP+FP)
recall = TP/(TP+FN)
f_score = 2*precision*recall/(precision+recall)
accuracy = train_res[train_res['pred']==train_res['actl']].shape[0]/train_res.shape[0]
print("acc is positive and unacc is negative")
print('precision: ' + str(precision) + ', recall: ' + str(recall) + ', f_score: ' + str(f_score))

# acc is positive and unacc is negative
TP = train_res_des['acc']['acc']
FN = train_res_des['unacc']['acc']
FP = train_res_des['acc']['unacc']
TN =  train_res_des['unacc']['unacc']
precision = TP/(TP+FP)
recall = TP/(TP+FN)
f_score = 2*precision*recall/(precision+recall)
accuracy = train_res[train_res['pred']==train_res['actl']].shape[0]/train_res.shape[0]
print("acc is positive and unacc is negative")
print('precision: ' + str(precision) + ', recall: ' + str(recall) + ', f_score: ' + str(f_score))

```

```python
# k has better to be odd so that the result will be not contain the same-target case
kn = [1, 3,5,7,9,11,15,21,25,31,35,41,45,51,60,71,81,91,101]
scores_1d = []
scores_2d = []
train_label = training['evaluation']
train_attr = training.drop('evaluation', axis=1)
train_input = pd.get_dummies(train_attr).values

from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score

for k in kn:
    scores_1d = []
    knn=KNeighborsClassifier(n_neighbors=k)
#     set the cv to 5 in default  
    for j in range(2,50):
        score = cross_val_score(knn,train_input,train_label,cv=j).mean() # average score of the cv
        scores_1d.append(score)
    scores_2d.append(scores_1d)
#     print(scores_2d)
```

```python
max_score = []
# print(scores_2d)
for i in range(len(scores_2d)):
    max_score.append(max(scores_2d[i]))
plt.plot(kn,max_score)
max(max_score)
```
