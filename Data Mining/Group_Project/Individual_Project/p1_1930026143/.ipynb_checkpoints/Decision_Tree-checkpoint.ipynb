# Implement the decision tree manually

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from treelib import tree
import warnings
warnings.filterwarnings("ignore")
from sklearn.model_selection import train_test_split
```

```python
training = pd.read_csv('./dataset/training.csv')
test = pd.read_csv('./dataset/test.csv')
train_train,train_test = train_test_split(training,test_size=0.25,random_state=17)
test_actl = list(train_test.iloc[:,-1].values)
test_attr = train_test.drop(['evaluation'], axis=1)
test_label = list(train_test.iloc[:,-1].values)
# print(test_label)
```

```python
# decide the featuer by gini coefficient
def decide_classify_attr(decide_attr, decide_label):
    entropy_attr = {}
    for i in decide_attr.keys():
        # return the best result key in the dict
        entropy_attr[i] = decide_label - decide_attr[i]
    # you should pay attention that it is a nestded dictionary
    decide_attr = max(entropy_attr, key=entropy_attr.get)
    return decide_attr



def entropy_decide(train_build):
    train_build_attr = train_build.columns[-1]
    train_build_label = train_build[train_build_attr].value_counts()
    build_label_dist = [train_build_label[0] / sum(train_build_label), train_build_label[1] / sum(train_build_label)]
    
    # get the distribution of the label in an attribution
    decide_label = entropy_cal(build_label_dist)

    decide_attr = {}
    train_build_len = train_build.shape[0]
    # get the gini information of the table
    for i in train_build.columns[:-1]:
        # Count up the items in each attributes 
        train_build_des = train_build[i].value_counts()
        entropy_res = 0
        
        for j in train_build_des.index:
            build_attr_type = train_build_des[j] / train_build_len
            train_build_label_des = train_build.iloc[:, -1][train_build[i] == j].value_counts()
            # If the length smaller than two, then gini is not increase 
            if len(train_build_label_des) < 2:
                entropy_res += 0
            else:
                attr_label_dist = [train_build_label_des[0] / sum(train_build_label_des), 
                                train_build_label_des[1] / sum(train_build_label_des)]
                entropy_res += build_attr_type * entropy_cal(attr_label_dist)
        decide_attr[i] = entropy_res
    return decide_label, decide_attr

def entropy_cal(x):
    sum = 0
    for i in x:
        sum += i ** 2
    return 1 - sum
```

```python
tree = Tree()
decision_making_order = []
tree.create_node(tag='root', identifier='root', data=0)
```

```python
def BuildDecisionTree(train_build, parent):
    if len(train_build.iloc[:, -1].unique()) == 1:
        tree.create_node(tag=train_build.iloc[:, -1].iloc[0],
                         parent=parent, 
                         identifier=parent + '_result',
                         data=train_build.iloc[:, -1].iloc[0])
        return
    elif len(train_build.columns) == 2:
        for i in train_build.iloc[:, 0].value_counts().index:
            tree.create_node(tag=train_build.columns[0] + '->' + i, parent=parent,
                             identifier=parent + train_build.columns[0] + '->' + i)
            dic = dict(train_build[train_build.iloc[:, 0] == i].iloc[:, -1].value_counts())
            tree.create_node(tag=max(dic, key=dic.get),
                             parent=parent + train_build.columns[0] + '->' + i,
                             identifier=parent + train_build.columns[0] + '->' + i + '_result',
                             data=max(dic, key=dic.get))
        return
    
    #                           Above part is the two cases of base cases
    #============================================================================================================
    #                                  Here's the recursive part
    
    # select the decision feature by the decide_classify_attr function
    decide_label, decide_attr = entropy_decide(train_build)
#     print([decide_label,decide_attr])
    decide_attr_key = decide_classify_attr(decide_attr, decide_label)
    if decide_attr[decide_attr_key] >= 0.4:
        decide_attr_dic = dict(train_build.iloc[:-1].value_counts())
                # connect the path  by the connection symbol -> as id
        # id can help us to find the parent and child of a node
        # Only the max one can add into the path preferentially
        tree.create_node(tag = max(decide_attr_dic, key = decide_attr_dic.get),
                         parent = parent + '->' + decide_attr_key,
                         identifier =parent + '->' + decide_attr_key+ '_result',
                         data = max(decide_attr_dic, key = decide_attr_dic.get))
        return
    for k in train_build[decide_attr_key].value_counts().index:
        if decide_attr_key not in decision_making_order:
            decision_making_order.append(decide_attr_key)
        new_data = train_build[train_build[decide_attr_key] == k].drop(columns=decide_attr_key, axis=1)
        # no need to add data just connect the attr and type as id
        tree.create_node(tag=decide_attr_key + '->' + k,
                         identifier=parent + '->' + decide_attr_key + '->' + k,
                         parent=parent)
        BuildDecisionTree(new_data, parent + '->' + decide_attr_key + '->' + k)
```

```python
BuildDecisionTree(train_train, 'root')
tree.show()
```

```python
pred_list = []
# Find the final data store in the tree structure by its edge
for i in range(len(test_attr)):
    searching_id = 'root'
    for j in decision_making_order:
        searching_id = searching_id + '->' + j + '->' + test_attr[j].iloc[i]

        if tree.get_node(searching_id) == None:
            #tree.get_node(searching_id + '->' + j)
            pred_list.append('acc')
            break
        elif tree.get_node(searching_id+'_result') == None:
            continue
        else:
            pred_list.append(tree.get_node(searching_id+'_result').data)
            break

# unacc is positive acse
def Score_unacc(pred_result, true_value):
    mistakes = 0
    acc_count = 0
    for w in range(len(pred_result)):
        if pred_result[w] != true_value[w]:
            mistakes += 1
        elif pred_result[w] == 'unacc':
            acc_count += 1
    accuracy = 1 - (mistakes / len(true_value))
    precise = acc_count / pred_result.count('unacc')
    recall = acc_count / true_value.count('unacc')
    f_measure = 2 * precise * recall / (precise + recall)
    score_dic = {'accuracy':accuracy, 'precise': precise, 'recall': recall, 'f_measure': f_measure}
    return score_dic
```

```python
print(Score_unacc(pred_list,test_label))
```

```python
# acc is positive acse
def Score_acc(pred_result, true_value):
    mistakes = 0
    acc_count = 0
    for w in range(len(pred_result)):
        if pred_result[w] != true_value[w]:
            mistakes += 1
        elif pred_result[w] == 'acc':
            acc_count += 1
    accuracy = 1 - (mistakes / len(true_value))
    precise = acc_count / pred_result.count('acc')
    recall = acc_count / true_value.count('acc')
    f_measure = 2 * precise * recall / (precise + recall)
    score_dic = {'accuracy':accuracy, 'precise': precise, 'recall': recall, 'f_measure': f_measure}
    return score_dic
print(Score_acc(pred_list, test_label))
```
