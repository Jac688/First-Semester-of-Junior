```python
import numpy as np
import pandas as pd
from catboost import CatBoostClassifier
from sklearn.model_selection import train_test_split
```

```python
training  = pd.read_csv("./dataset/training.csv")
test = pd.read_csv("./dataset/test.csv")
train_attr, test_attr, train_target, test_target = train_test_split(training.iloc[:,:-1],training.iloc[:,-1],test_size=0.2 , random_state=3333)
```

```python
test.iloc[:,:-1]
```

```python
decision_feature = np.where(train_attr.dtypes != np.float)[0]
# The loss function is crossing entroy
# by default, the deepth is 16 which can effect the speed of convergence 
# With the increase of time, the convergence speed gradually slows down, and the final error is almost unchanged
catboost = CatBoostClassifier(iterations=100, depth=10,
                              cat_features=decision_feature,learning_rate=0.2, loss_function='Logloss',
                              logging_level='Verbose')
```

```python
catboost.fit(train_attr,train_target,eval_set=(test_attr, test_target),plot=True)
```

```python
print(catboost.get_best_score())
```

```python
res_pred = catboost.predict(test).tolist()
res_des = pd.Series(res_pred).value_counts()
res_des[0]/res_des.sum(), res_des[1]/res_des.sum()
res_pred
```

```python
res_pred = catboost.predict(training.drop(['evaluation'], axis=1)).tolist()
res_ser = pd.Series(res_pred)
```

```python
train_res = pd.DataFrame([res_ser,training['evaluation']]).T
train_res.columns = ['pred', 'actl']
train_res_des = train_res.value_counts()


# score with acc is negative and unacc is positive
TP = train_res_des['unacc']['unacc']
FN = train_res_des['acc']['unacc']
FP = train_res_des['unacc']['acc']
TN =  train_res_des['acc']['acc']
precision = TP/(TP+FP)
recall = TP/(TP+FN)
f_score = 2*precision*recall/(precision+recall)
accuracy = train_res[train_res['pred']==train_res['actl']].shape[0]/train_res.shape[0]
print("acc is positive and unacc is negative")
print('precision: ' + str(precision) + ', recall: ' + str(recall) + ', f_score: ' + str(f_score))

# acc is positive and unacc is negative
TP = train_res_des['acc']['acc']
FN = train_res_des['unacc']['acc']
FP = train_res_des['acc']['unacc']
TN =  train_res_des['unacc']['unacc']
precision = TP/(TP+FP)
recall = TP/(TP+FN)
f_score = 2*precision*recall/(precision+recall)
accuracy = train_res[train_res['pred']==train_res['actl']].shape[0]/train_res.shape[0]
print("acc is positive and unacc is negative")
print('precision: ' + str(precision) + ', recall: ' + str(recall) + ', f_score: ' + str(f_score))
```
