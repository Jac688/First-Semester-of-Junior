```python
def bayes(training, test):
    # The dictionary query system 
    # cerate a list to store information for the atrributes
    list_attr_3d = []
    list_attr_2d = []
    list_attr_1d = []
    list_res = []
    res = training.iloc[:,-1]
#     print(len(res))
#     transform the data to statistics 
    res_des=res.value_counts()
    res_len=len(res_des)
    list_attr = []
    list_attr2 = []
    list_attr3 = []
    list_test = []
    list_test2 = []
    for i in range(0,res_len):
#         calcalate the prior probability 
        P_C = (res_des[i]+1)/(len(res)+res_len)
        list_res.append(P_C)
#         select the conditionby the training
        res_con = training[training.iloc[:, -1]==res_des.index[i]]
        list_attr_2d = []
        list_attr2 = []
        attr_len = 0
#         put the data into the diectory
        for j in range(0,len(res_con.columns)-1):
            attr = res_con.iloc[:,j]
            attr_des = attr.value_counts()
            attr_len = len(attr_des)
            
            list_attr_1d = []
            list_attr = []
            
            for k in range(0, attr_len):
                # introduce Laplacian operation
                # training set is not very large, there may be situations that have not appeared in the training set in the test set
                P_C_con = (attr_des[k]+1) / (res_des[i]+attr_len)
                list_attr.append(attr_des.index[k])
                list_attr_1d.append(P_C_con)
            list_attr_2d.append(list_attr_1d)
            list_attr2.append(list_attr)
        # For unacc,acc, divided into two categories for each attribute
        # For each attribute, divided into each label
        list_attr_3d.append(list_attr_2d) 
        list_attr3.append(list_attr2)
        
#         print(list_attr_3d)
#         print(list_attr3)
        list_test = []
        for m in range(0, len(test.index)):
            mult = 1
            # query the dictionary, find out the test case in the training one to aviod error
            for n in range(0, len(test.columns)):
                # If there is no case in the training, just times 1 in numerator 
                if test.iloc[m,n] not in list_attr3[i][n]:
                    mult = mult*1/(res_des[i]+attr_len)
                else:
                    # query 
                    idx = list_attr3[i][n].index(test.iloc[m,n])
                    mult = mult*list_attr_3d[i][n][idx]
            mult_res = mult*list_res[i]
            list_test.append(mult_res)
        list_test2.append(list_test)
#     generate a dataframe to show the result
    df = pd.DataFrame(list_test2).T
    df.columns = list(res_des.index)
    print(list(res_des.index))
    df['pred'] = np.max(df, axis = 1)
    pred_res = df.copy()
    for i in range(0, len(df.index)):
        for j in range(0, len(df.columns)-1):
            if pred_res['pred'][i] == pred_res.iat[i, j]:
                pred_res['pred'][i] = pred_res.columns[j]
    return pred_res

# print(res[0][:5])
# print(res[1][:5])
# len(res[0]), len(res[1])

```

```python
warnings.filterwarnings('ignore')
# predict the outcome and  
bayes_res = bayes(training, test)
print(bayes_res)
unacc_proportion = len(bayes_res[bayes_res['pred']=='unacc'].index)/len(bayes_res.index)
acc_proportion = len(bayes_res[bayes_res['pred']=='acc'].index)/len(bayes_res.index)
print(unacc_proportion, acc_proportion)
```

```python
# predict the results of the tests
warnings.filterwarnings('ignore')
bayes_res_train = bayes(training, training.drop(['evaluation'], axis=1))
unacc_proportion_train = len(bayes_res_train[bayes_res_train['pred']=='unacc'].index)/len(bayes_res_train.index)
acc_proportion_train = len(bayes_res_train[bayes_res_train['pred']=='acc'].index)/len(bayes_res_train.index)
print(unacc_proportion_train, acc_proportion_train)
warnings.filterwarnings('ignore')
bayes_res_train
```

```python
def score():
    train_res = pd.DataFrame([bayes_res_train['pred'], training['evaluation']]).T
    train_res.columns = ['pred', 'actl']
    train_pred_unacc = train_res[train_res['pred']=='unacc']
    train_FP = train_pred_unacc[train_pred_unacc['actl']=='acc'].shape[0]
    train_TP = train_pred_unacc[train_pred_unacc['actl']=='unacc'].shape[0]
    # print('train_FP' + ': ' + str(train_FP), 'train_TP' + ': ' + str(train_TP))

    train_pred_acc = train_res[train_res['pred']=='acc']
    train_FN = train_pred_acc[train_pred_acc['actl']=='acc'].shape[0]
    train_TN = train_pred_acc[train_pred_acc['actl']=='unacc'].shape[0]
    # print('train_FN' + ': ' + str(train_FN), 'train_TN' + ': ' + str(train_TN))
    precision = train_TP / (train_TP+train_FP)
    recall = train_TP / (train_TP+train_FN)
    f_score = 2*precision*recall/(precision+recall)
    print("unacc is positive and acc is negative")
    print('precision: ' + str(precision) + ', recall: ' + str(recall) + ', f_score: ' + str(f_score))
    print("===========================================")
    
    train_pred_unacc = train_res[train_res['pred']=='acc']
    train_FP = train_pred_unacc[train_pred_unacc['actl']=='unacc'].shape[0]
    train_TP = train_pred_unacc[train_pred_unacc['actl']=='acc'].shape[0]
    # print('train_FP' + ': ' + str(train_FP), 'train_TP' + ': ' + str(train_TP))

    train_pred_acc = train_res[train_res['pred']=='unacc']
    train_FN = train_pred_acc[train_pred_acc['actl']=='acc'].shape[0]
    train_TN = train_pred_acc[train_pred_acc['actl']=='unacc'].shape[0]
    # print('train_FN' + ': ' + str(train_FN), 'train_TN' + ': ' + str(train_TN))
    precision = train_TP / (train_TP+train_FP)
    recall = train_TP / (train_TP+train_FN)
    f_score = 2*precision*recall/(precision+recall)
    print("acc is positive and unacc is negative")
    print('precision: ' + str(precision) + ', recall: ' + str(recall) + ', f_score: ' + str(f_score))
score()
```
